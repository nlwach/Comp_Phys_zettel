```python
def I_j_plus_one(a, I_j, c, m):
    return (a*I_j+c)%m

def norm(I, m):
    return I/(m-1)

def calc(m, a, c, steps, initial=40):
    I = norm(initial, m)
    rand_num = [I]
    for i in range(steps):
        I = I_j_plus_one(a, I, c, m)
        rand_num.append(norm(I, m))
    return rand_num
        
```

```python
def dice_calc(steps, size=10):
    result = []
    for n in range(steps):
        dice_float = np.array(calc(6075, 106, 1283, size, initial=n))*6
        dice_int = [int(np.ceil(dice)) for dice in list(dice_float)]
        dice_sum = np.sum(np.array(dice_int))
        result.append(dice_sum)
    return result
        
```
