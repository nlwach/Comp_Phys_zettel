```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
def I_j_plus_one(a, I_j, c, m):
    return (a*I_j+c)%m

def norm(I, m):
    return I/(m-1)

def calc(m, a, c, steps, initial=40):
    I = norm(initial, m)
    rand_num = [I]
    for i in range(steps):
        I = I_j_plus_one(a, I, c, m)
        rand_num.append(norm(I, m))
    return rand_num
        
```

```python
m = 106
n = calc(6075, 106, 1283, 10000)
```

```python
plt.hist(n)
```

```python
s = calc(6075, 106, 1283, 1000, initial=40)
r = calc(6075, 106, 1283, 1000, initial=20)

plt.scatter(r, s)
```

```python
I_j = calc(6075, 106, 1283, 1000, initial=40)
I_j_plus_one= I_j
del I_j_plus_one[0]
del I_j[-1]
```

```python
plt.plot(I_j, I_j_plus_one)
```
