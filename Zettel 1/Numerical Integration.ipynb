```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## a) Plot of the Integrand

```python
def func (n,a):
    return x**n/(x+a)

a = 5
n = np.concatenate([np.array([1,5]),np.arange(10,60,10)]) #create an array of the given n
x = np.linspace(0,1,1000)

plt.figure(figsize=(11,8))

for i in range(0,len(n)):
    plt.plot(x,func(n[i],a), label='n= %f'%((n[i]))) #plot the function for different n
    
plt.legend()

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of the Integrand for different n')
```

## b) Iteration

```python
def y_up(a,n,y):
    return 1/n - a*y #y_n out of y_n-1

def y_down(a,n,y):
    return 1/a * (1/n - y) #y_n-a out of y_n

def iteration(a,n0,n1,y0):
    if n0 == n1: #if n0=n1 there is no iteration
        print('n0 and n1 are the same number')
    if n0 < n1: #iterate upwards if n0<n1
        for i in range(n0+1,n1+1): 
            y = y_up(a,i,y0)
            y0 = y
    if n0 > n1: #iterate downwards if n1<n0
        for i in range(n0,n1,-1):
            yy = y_down(a,i,y0)
            y0 = yy
    print(y0)
        
```

## c) Experiment with the Iteration

```python
iteration(5,0,30,np.log((1+5)/5)) #a=5,n0=0,n1=30,y0=ln(1+a/a) 
iteration(5,50,30,1) #a=5,n0=50,n1=30,y0=1
```
